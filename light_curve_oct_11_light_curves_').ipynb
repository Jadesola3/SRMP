```python
df =pd.read_csv('KIC_2157356.txt')
plt.plot (df['time'], df['flux'], '.')
plt.xlabel('time[days]')
plt.ylabel('flux')
```

```python
from astropy.timeseries import LombScargle
```

```python
import numpy as np
frequency = np.linspace(1/90, 1,1000)
```

1/90= period, 1= min

```python
frequency
```

```python
power = LombScargle(df['time'], df['flux']).power(frequency)
```

```python
plt.plot(1./frequency, power)
```
